```python
import pandas as pd
import numpy as np
```

```python
data = {
    "free":[0,1,1,2],
    "win":[0,0,1,1],
    "spam":[0,0,1,1]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
X = df[["free","win"]]
y = df["spam"]
```

```python
X
```

```python
y
```

```python
w1 = 1.5   # weight for 'free'
w2 = 2.0   # weight for 'win'
b  = -2.0  # bias
```

```python
# Linear Equation(z)
df["z"] = w1*df["free"] + w2*df["win"]+b
df
```

```python
# Sigmmoid function
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
df["probability"] = sigmoid(df["z"])
df
```

```python
# Predication Using Threshold
df["prediction"] = (df["probability"] >= 0.5).astype(int)
```
